## zbiÃ³r **square-simple**

```python
import pandas as pd 

data = pd.read_csv("C:\\Users\\natal\\Desktop\\STUDIA\\SEM_6\\MIOWAD\\dane\\regression\\square-simple-test.csv")
X = data.iloc[:, 1].values.reshape(-1, 1)
y_true = data.iloc[:, 2].values.reshape(-1, 1)
```

### architektura: 1 -> 5 -> 1

```python
from NN1.MLP_nobackprop import MLP
import numpy as np
import matplotlib.pyplot as plt

mlp = MLP([1,5,1])

# W0 = np.ones((1,5))
# b0 = np.zeros((1,5))
# W1 = np.ones((5,1))
# b1 = np.zeros((1,1))

# MSE = 163
# W0 = [[-4.926272, -9.241243, -25.235859, -6.9100876, -31.939518]]
# b0 = [[-5.3270583, 33.000164, 44.606815, 4.841127, 46.014847]]
# W1 = [[205.70866], [-35.166595], [-90.29393], [-124.01518], [-95.28988]]
# b1 = [[218.56583]]

# MSE = 6
W0 = [[0.217, 0.653, -0.746, 3, 2]]
b0 = [[1.816, 2.195, 2.649, 0.3, 1]]
W1 = [[4168], [-3898], [-2571], [3], [-4]]
b1 = [[2195.4]]

mlp.set_weights_and_biases(0, W0, b0)
mlp.set_weights_and_biases(1, W1, b1)

y_pred = mlp.predict(X)
mse = mlp.mse(y_true, y_pred)

print("MSE = ", mse)

plt.scatter(X, y_true, label="True Data")
plt.scatter(X, y_pred, label="Prediction")
plt.title(f'Prediction vs True Data (MSE: {mse:.4f})')
plt.legend()
plt.show()
```
